1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

url = 'https://raw.githubusercontent.com/dsvitorchaves/EBAC/main/previsao_de_renda.csv'

dados = pd.read_csv(url)
dados.head()
````

````python
# Verifica se 'Unnamed: 0' está presente e remove a coluna
if 'Unnamed: 0' in dados.columns:
    dados = dados.drop('Unnamed: 0', axis=1)

dados.head()
````

````python
# Converter 'data_ref' para datetime
dados['data_ref'] = pd.to_datetime(dados['data_ref'])

# Extrair o ano da variável 'data_ref' para facilitar a agregação temporal
dados['ano'] = dados['data_ref'].dt.year
````

````python
# Verificar os valores únicos na coluna 'posse_de_veiculo'
print(dados['posse_de_veiculo'].unique())

# Se os valores forem booleanos, converta-os para strings
if dados['posse_de_veiculo'].dtype == 'bool':
    dados['posse_de_veiculo'] = dados['posse_de_veiculo'].map({True: 'Sim', False: 'Não'})

plt.figure(figsize=(12, 6))
sns.countplot(data=dados, x='ano', hue='posse_de_veiculo')
plt.title('Distribuição de Posse de Veículo ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.legend(title='Possui Veículo')
plt.show()
````

````python
# Verificar os valores únicos na coluna 'posse_de_imovel'
print(dados['posse_de_imovel'].unique())

# Se os valores forem booleanos, converta-os para strings
if dados['posse_de_imovel'].dtype == 'bool':
    dados['posse_de_imovel'] = dados['posse_de_imovel'].map({True: 'Sim', False: 'Não'})

plt.figure(figsize=(12, 6))
sns.countplot(data=dados, x='ano', hue='posse_de_imovel')
plt.title('Distribuição de Posse de Imóvel ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.legend(title='Possui Imóvel')
plt.show()
````

````python
plt.figure(figsize=(12, 6))
sns.countplot(data=dados, x='ano', hue='tipo_renda')
plt.title('Distribuição do Tipo de Renda ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Renda')
plt.show()
````

````python
plt.figure(figsize=(12, 6))
sns.countplot(data=dados, x='ano', hue='educacao')
plt.title('Distribuição do Nível de Educação ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.legend(title='Nível de Educação')
plt.show()
````

````python
plt.figure(figsize=(12, 6))
sns.countplot(data=dados, x='ano', hue='estado_civil')
plt.title('Distribuição do Estado Civil ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.legend(title='Estado Civil')
plt.show()
````

````python
plt.figure(figsize=(12, 6))
sns.countplot(data=dados, x='ano', hue='tipo_residencia')
plt.title('Distribuição do Tipo de Residência ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Residência')
plt.show()
````

Ao analisar os gráficos de barras que mostram a contagem das variáveis qualitativas ao longo dos anos, observamos uma diminuição significativa em relação ao ano anterior para várias dessas variáveis.

- Posse de Veículo: A contagem de pessoas com posse de veículo mostrou uma redução em comparação com o ano anterior, indicando uma possível diminuição na aquisição ou manutenção de veículos.

- Posse de Imóvel: A contagem de pessoas com posse de imóvel também apresentou uma queda, sugerindo uma menor taxa de aquisição ou manutenção de propriedades residenciais.

- Tipo de Renda: Houve uma diminuição na quantidade de pessoas em cada categoria de tipo de renda, o que pode refletir mudanças no mercado de trabalho ou na forma como os indivíduos são remunerados.

- Nível de Educação: A contagem de pessoas em diferentes níveis de educação mostrou uma redução, o que pode indicar uma mudança na frequência de novos graduados ou na forma como a educação é registrada.

- Estado Civil: Observou-se uma diminuição na contagem de indivíduos em cada estado civil, sugerindo possíveis mudanças nos padrões de casamento, divórcio ou outros estados civis.

- Tipo de Residência: A contagem das diferentes categorias de tipo de residência apresentou uma redução, possivelmente refletindo mudanças nas preferências residenciais ou nas condições do mercado imobiliário.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
# Função para converter booleanos em strings para variáveis qualitativas
def convert_boolean_to_string(df, column):
    if df[column].dtype == 'bool':
        df[column] = df[column].map({True: 'Sim', False: 'Não'})

# Converter booleanos para as variáveis qualitativas
convert_boolean_to_string(dados, 'posse_de_veiculo')
convert_boolean_to_string(dados, 'posse_de_imovel')
````

````python
plt.figure(figsize=(12, 6))
sns.barplot(data=dados, x='ano', y='renda', hue='posse_de_veiculo', ci=None)
plt.title('Média da Renda por Posse de Veículo ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Média da Renda')
plt.legend(title='Possui Veículo')
plt.show()
````

- A análise dos gráficos de barras acima mostra que a média da renda por posse de veículo revelou uma tendência notável de aumento ao longo do tempo. Comparando os dados de um ano para o outro, observamos que a média da renda para pessoas com posse de veículo aumentou consistentemente.

````python
plt.figure(figsize=(12, 6))
sns.barplot(data=dados, x='ano', y='renda', hue='posse_de_imovel', ci=None)
plt.title('Média da Renda por Posse de Imóvel ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Média da Renda')
plt.legend(title='Possui Imóvel')
plt.show()
````
